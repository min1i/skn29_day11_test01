```python
# 중복데이터 확인(개수)

import pandas as pd

df = pd.read_csv('data/messy_data.csv')
df.duplicated().sum()
```

```python
# 중복데이터(내용)
d_index = df.duplicated(keep=False)
df[d_index].sort_values(by=['id'])
```

```python
# id는 중복되면 안됨
id_d = df.duplicated(subset=['id'])
df[id_d]
```
